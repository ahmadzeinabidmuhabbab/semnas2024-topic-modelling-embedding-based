```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib.ticker import MultipleLocator
```

```python
df_cv = pd.read_csv('result/cv_scores_rev.csv',index_col=0)
df_td = pd.read_csv('result/td_scores_rev.csv',index_col=0)
df_npmi = pd.read_csv('result/npmi_scores_rev.csv',index_col=0)
# df_cv = df_cv.drop(columns=['Unnamed: 0.5','Unnamed: 0.4','Unnamed: 0.3','Unnamed: 0.2','Unnamed: 0.1','Unnamed: 0'])
# df_td = df_td.drop(columns=['Unnamed: 0.5','Unnamed: 0.4','Unnamed: 0.3','Unnamed: 0.2','Unnamed: 0.1','Unnamed: 0'])
# df_npmi = df_npmi.drop(columns=['Unnamed: 0.5','Unnamed: 0.4','Unnamed: 0.3','Unnamed: 0.2','Unnamed: 0.1','Unnamed: 0'])
```

```python
n_cluster = list(range(2,21))
df_cv['n_cluster'] = n_cluster
df_td['n_cluster'] = n_cluster
df_npmi['n_cluster'] = n_cluster
```

```python
df_cv.to_csv('result/cv_scores.csv')
df_td.to_csv('result/td_scores.csv')
df_npmi.to_csv('result/npmi_scores.csv')
```

```python
df_cv = df_cv[df_cv['n_cluster']<=10]
df_td = df_td[df_td['n_cluster']<=10]
df_npmi = df_npmi[df_npmi['n_cluster']<=10]
```

```python
dfl = pd.melt(df_cv, ['n_cluster'])
dfl.columns = ['n_cluster', 'model', 'CV']
# dfl = dfl[dfl['model']!='Unnamed: 0']
sns.lineplot(data=dfl, x='n_cluster', y='CV', hue='model', )

# Menyesuaikan axis agar berupa integer
plt.gca().xaxis.set_major_locator(MultipleLocator(1))  # Setiap 1 langkah di sumbu x

# Mengubah judul sumbu x
plt.xlabel('Banyak Topik')

# Mengubah judul legend dan isi legend
plt.legend(title='Model Embedding')

# Menampilkan plot
plt.show()
```

```python
dfl = pd.melt(df_td, ['n_cluster'])
sns.lineplot(data=dfl, x='n_cluster', y='value', hue='variable')
# Menyesuaikan axis agar berupa integer
plt.gca().xaxis.set_major_locator(MultipleLocator(1))  # Setiap 1 langkah di sumbu x

# Mengubah judul sumbu x dan y
plt.xlabel('Banyak Topik')
plt.ylabel('TD')

# Mengubah judul legend dan isi legend
plt.legend(title='Model Embedding')

# Menampilkan plot
plt.show()
```

```python
dfl = pd.melt(df_npmi, ['n_cluster'])
sns.lineplot(data=dfl, x='n_cluster', y='value', hue='variable')
# Menyesuaikan axis agar berupa integer
plt.gca().xaxis.set_major_locator(MultipleLocator(1))  # Setiap 1 langkah di sumbu x

# Mengubah judul sumbu x dan y
plt.xlabel('Banyak Topik')
plt.ylabel('NPMI')

# Mengubah judul legend dan isi legend
plt.legend(title='Model Embedding')

# Menampilkan plot
plt.show()
```
